```python
import urllib2
from scrapy.selector import Selector
from os import listdir
from os.path import isfile, join
```

```python
def dlpage(url, targetname):
    """pass url and file name, downloads url page to this file"""
    file(targetname, "wb").write(urllib2.urlopen(url).read())
    
def getFiles(path):
    """Function to return a list of all files within a folder"""
    files = [ f for f in listdir(path) if isfile(join(path,f)) and f[0] != '.' ]
    return files
```

```python
base = 'DH2012/source/'
searchpages = [base + x for x in getFiles(base)]
#print searchpages
```

```python
# Extract the abstract URLs from the search pages
# and download those abstract files
```

```python
foundlinks = []

for f in searchpages:
    with open(f, 'rt') as text:
        body = text.read()
        pages = Selector(text=body).xpath('//a[starts-with(@href,\'view?docId=tei/\')]/@href').extract()
        foundlinks += pages
```
