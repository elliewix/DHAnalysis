```python
import urllib2
from scrapy.selector import Selector
from os import listdir
from os.path import isfile, join
```

```python
def dlpage(url, targetname):
    """pass url and file name, downloads url page to this file"""
    file(targetname, "wb").write(urllib2.urlopen(url).read())
    
def getFiles(path):
    """Function to return a list of all files within a folder"""
    files = [ f for f in listdir(path) if isfile(join(path,f)) and f[0] != '.' ]
    return files
```

```python
base = 'DH2013/abstracts/'

files = [base + f for f in getFiles(base)]
```

```python
removethings = ['<em>','</em>','<b>','</b>','<i>','</i>']

for ab in files:
    print ab
    name = ab.split('/')[-1].split('.')[0]
    print name
    with open(ab,'rb') as abfile:
        body = abfile.read()
    for each in removethings:
        body = body.replace(each, '')
    text = "\n".join(Selector(text=body).xpath('//text//text()').extract())
    print text[:300]
    with open('DH2013/textfiles/' + name + '.txt', 'wt') as abwrite:
        abwrite.write(text.encode('utf8'))
```
